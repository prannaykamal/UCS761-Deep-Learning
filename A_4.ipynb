```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/multiple_linear_regression_dataset.csv")

data.head()
```

```python
X = data[["age","experience"]].values
y = data["income"].values

n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0
```

```python
def predict(X,w,b):
    return X.dot(w) + b

def mean_squared_error(y,y_hat):
    return ((y_hat - y)**2).mean()

def compute_gradients(X,y,y_hat):
    N = len(y)
    dw = (2/N) * X.T.dot(y_hat - y)
    db = (2/N) * (y_hat - y).sum()
    return dw,db

def update_parameters(w,b,dw,db,lr):
    w = w - lr*dw
    b = b - lr*db
    return w,b
```

```python
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X,w,b)
    loss = mean_squared_error(y,y_hat)
    dw,db = compute_gradients(X,y,y_hat)
    w,b = update_parameters(w,b,dw,db,lr)
    if epoch % 100 == 0:
        print(epoch,loss)

print("Weights:",w)
print("Bias:",b)
```

```python
new_candidate = np.array([68,4.5])
predicted_salary = new_candidate.dot(w) + b
print(predicted_salary)
```
