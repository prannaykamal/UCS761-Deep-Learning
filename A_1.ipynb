```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])

gates = {
    "AND":  np.array([0,0,0,1]),
    "OR":   np.array([0,1,1,1]),
    "NAND": np.array([1,1,1,0]),
    "NOR":  np.array([1,0,0,0]),
    "XOR":  np.array([0,1,1,0])
}

def train(X,y,lr=0.1,epochs=20):
    w=np.zeros(2)
    b=0
    for _ in range(epochs):
        for i in range(len(X)):
            z=np.dot(X[i],w)+b
            yhat=1 if z>=0 else 0
            e=y[i]-yhat
            w=w+lr*e*X[i]
            b=b+lr*e
    return w,b

for name,y in gates.items():
    w,b=train(X,y)
    print(name)
    print("weights:",w)
    print("bias:",b)
    preds=[1 if np.dot(x,w)+b>=0 else 0 for x in X]
    print("predictions:",preds,"\n")
```
