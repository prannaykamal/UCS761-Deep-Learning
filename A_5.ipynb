```python
import pandas as pd
import numpy as np
```

```python
cols=["Sex","Length","Diameter","Height","WholeWeight","ShuckedWeight","VisceraWeight","ShellWeight","Rings"]
data=pd.read_csv("abalone.data",header=None,names=cols)

print(len(data))
print(data.columns)
print(data.head())
```

```python
y=(data["Rings"]+1.5).values.reshape(-1,1)

x1=(data["Length"]*data["Height"]).values
x2=data["Diameter"].values
x3=data["WholeWeight"].values

X=np.column_stack((x1,x2,x3))
```

```python
N=len(X)
split=int(0.8*N)

X_train=X[:split]
X_test=X[split:]
y_train=y[:split]
y_test=y[split:]

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
mean=X_train.mean(axis=0)
std=X_train.std(axis=0)

X_train=(X_train-mean)/std
X_test=(X_test-mean)/std

d=X_train.shape[1]

w=np.random.randn(d,1)*0.01
b=0.0
```

```python
def forward(X,w,b):
    y_hat=X@w+b
    print(X.shape,w.shape,type(b),y_hat.shape)
    return y_hat

def mse(y,y_hat):
    return np.mean((y-y_hat)**2)

def mae(y,y_hat):
    return np.mean(np.abs(y-y_hat))

def grad_w(X,y,y_hat):
    N=len(y)
    return (2/N)*X.T@(y_hat-y)

def grad_b(y,y_hat):
    N=len(y)
    return (2/N)*np.sum(y_hat-y)
```

```python
lr=0.01
epochs=200

for epoch in range(epochs):
    y_hat=forward(X_train,w,b)
    loss=mse(y_train,y_hat)
    dw=grad_w(X_train,y_train,y_hat)
    db=grad_b(y_train,y_hat)
    w=w-lr*dw
    b=b-lr*db
    if epoch%20==0:
        print(epoch,loss)

y_pred=forward(X_test,w,b)

test_mse=mse(y_test,y_pred)
test_mae=mae(y_test,y_pred)

print("Test MSE:",test_mse)
print("Test MAE:",test_mae)
```
